```python
import pandas as pd
```

```python
## 검색어 정리해서 주소 위경도는 다른 팀원이 주소 결합
```

```python
# 숙소주소 정리된 데이터
df_addr = pd.read_excel('data/숙소주소_결합본.xlsx')
df_addr.head()
```

```python
df_lng = pd.read_excel('data/숙소주소_위경도.xlsx')
df_lng.head()
```

```python
df_addr = df_addr.merge(df_lng, how = 'left', on=['검색어'])
df_addr.head()
```

```python
# 숙소 주소가 NaN인경우 -> 60개
df_addr['위도'].isna().sum()
```

```python
df_addr.columns
```

```python
df_base = df_addr[['title', 'day', 'price', 'rating', 'reviewNum', 'location', '검색어', '주소1_x', '위도', '경도']]
df_base.head()
```

```python
# 컬럼명 변경
df_base.rename(columns= {'검색어' : 'pension_keyword', '주소1_x' : 'addr', '위도' : 'lat', '경도' : 'lng'}, inplace= True)
df_base.head(3)
```

```python
# 주소가 확인되는 숙소 선택 - 96개 선택
df_base = df_base[df_base['lat'].notna()]
len(df_base)
```

```python
# 이름이 중복되는 컬럼 존재
df_base.duplicated('title').sum()
```

```python
df_base[df_base.duplicated('title')]
```

```python
# '[제주/구좌] 덕천씨엘빌'가 4개로 등록되어서 하나로 바꾼다.
df_base[df_base['title'] == '[제주/구좌] 덕천씨엘빌'] 
```

```python
# 두개의 정보를 하나로 합치기
new_rating = round(((4.9 * 11) + (4.7 * 9)) / 20, 1)
temp = pd.DataFrame([['[제주/구좌] 덕천씨엘빌', '29박', 2871000, new_rating, 20, '구좌읍', '[제주/구좌]덕천씨엘빌', '제주 제주시 구좌읍 덕평로 121-101', '33.495082', '126.753155']])
temp.columns = df_base.columns
temp
```

```python
# 총 93개의 케이스 존재
temp2 = df_base[df_base['title'] != '[제주/구좌] 덕천씨엘빌'] # 중복된거 제거
df_base = pd.concat([temp, temp2]) # 중복된 항 처리
df_base.reset_index(drop = True, inplace= True)
df_base
```

```python
# 총93개의 case
len(df_base)
```

```python
df_base.to_csv('./data/mmention_base_20240118.csv', index=False)
```

### review데이터의 ammen (시설정보) 추가

```python
df_detail =pd.read_csv('./data/mmention_review_20240118.csv')
df_detail.head(3)
```

```python
df_ammen = df_detail[['title', 'ammen']]
df_ammen.head(3)
```

```python
# ammen은 숙소별로 1개씩만 필요
df_ammen.drop_duplicates(inplace= True)
df_ammen
```

# 두 데이터 프레임 합치기

```python
df = pd.merge(df_base, df_ammen, how='left', on = 'title')
df
```

```python
# 1개 항목 사라짐
df.drop_duplicates('title', inplace= True)
df.reset_index(drop=True, inplace= True)
len(df)
```

```python
# 숙소 익명화를 위해 ID 부여
pensionID = [f'pension{x + 1}' for x in range(len(df))]
df['pensionID'] = pensionID
df
```

```python
# ammen 항 전처리
df['ammen'] = df['ammen'].apply(lambda x: str(x).replace("'", "").replace("[", "").replace("]", ""))
df['ammen']
```

```python
# ammen 항 5개 부분임으로 5 개 영역으로 분리
df_temp = df['ammen'].str.split(",", expand=True)
df_temp.columns = ['ammen1', 'ammen2', 'ammen3', 'ammen4', 'ammen5']
df_temp.head(5)
```

```python
df_final = pd.concat([df, df_temp], axis=1)
df_final.head()
```

```python
# 숙소시설 정보가 없는 경우 제외
print(len(df))
df = df[df['ammen'] != 'nan']
print('제외후', len(df))
```
