## 제주도 한달살기, 한달살이 블로그 데이터 전처리1



```python
import pandas as pd
```

```python
df1 = pd.read_csv('./data/jeje_month1.csv')
df2 = pd.read_csv('./data/jeje_month2.csv')
```

```python
df1.head()
```

```python
df2.head()
```

```python
### 수집오류로 blog_text가 윗 행과 같게 수집되는 경우 존재
df1['blog_text2'] = df1['blog_text'].shift(1)
df1.head(5)
```

```python
# data1
len(df1[df1['blog_text'] == df1['blog_text2']])
# 총 75개의 행이 잘못 수집됨
# 다시 수집하기에는 시간이 문제
# 1050개 중에서 75개 제외
```

```python
df2['blog_text2'] = df2['blog_text'].shift(1)
len(df2[df2['blog_text'] == df2['blog_text2']])
# 총 88개의 행이 잘못 수집됨
# 다시 수집하기에는 시간이 문제
# 1050개 중에서 88개 제외
```

```python
df1 = df1[df1['blog_text'] != df1['blog_text2']]
df1['key'] = '한달살기'
df1.drop(columns='blog_text2', inplace=True)

df2 = df2[df2['blog_text'] != df2['blog_text2']]
df2['key'] = '한달살이'
df2.drop(columns='blog_text2', inplace=True)

df = pd.concat([df1, df2])
df
```

```python
df.to_csv('./data/naver_month.csv', index=False)
```
