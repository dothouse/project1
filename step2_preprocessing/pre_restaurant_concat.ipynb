```python
import pandas as pd
import numpy as np
import glob
```

```python
output = glob.glob('con_data/consume_data/*.csv')
output
```

```python
dt_list = [] # 빈 리스트에 con_data에 있는 파일을 모두 넣기
for x in sorted(glob.glob('con_data/consume_data/*.csv')):
    temp = pd.read_csv(x)
    dt_list.append(temp) #월 별로 나뉘어져 있던 자료들을 하나로 통합하기 위함

df = pd.concat(dt_list,axis=0) # 리스트를 판다스로 변환

df.shape # df의 행과 열을 볼 수 있음
```

```python
df = df.drop(['ANALS_YM','SIGNGU_NM','AREA_NM','VARTION_RT','RANK_CO','GID_ID'],axis=1) # 필요없는 컬럼 삭제
df.head()
```

```python
df.to_csv('con_data/consume.csv',index=False)
```
