```python
location_data=[[33.452316, 126.923895],
               [33.405634, 126.904469],
               [33.325612, 126.839372],
               [33.277817, 126.719605],
               [33.259451, 126.623882],
               [33.247405, 126.558657],
               [33.248993, 126.508219],
               [33.243505, 126.458519],
               [33.236981, 126.360720],
               [33.240186, 126.335446],
               [33.173513, 126.268679],
               [33.218657, 126.252418],
               [33.270937, 126.238749],
               [33.323733, 126.167259],
               [33.333691, 126.255967],
               [33.415731, 126.261589],
               [33.466908, 126.337848],
               [33.460503, 126.439258],
               [33.512571, 126.524320],
               [33.541027, 126.640726],
               [33.557664, 126.744866],
               [33.523610, 126.863426],
               [33.476824, 126.895518]
              ]

folium.Polygon(locations=location_data, tooltip='PolyLine').add_to(map)
```

```python
colors = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0']
```

```python
from itertools import cycle
```

```python
color_cycle = cycle(colors)
for i in range(len(location_data) - 1):
    coords = [location_data[i], location_data[i + 1]]
    color = next(color_cycle)
    folium.PolyLine(locations=coords, color=color, weight=5, tooltip=f'Polyline {i+1}').add_to(map)
```

```python
for coords, color in zip(location_data, colors):
    folium.CircleMarker(location=coords, radius=5, color=color, fill=True, fill_color=color).add_to(map)
```

```python
# 지도 생성
map = folium.Map(location=[df.위도.mean(), df.경도.mean()], zoom_start=10)
for i in df.index:
    folium.Marker(
        location=[df.위도[i], df.경도[i]],
        tooltip=df.코스시점[i],
        popup = folium.Popup(df.코스시점주소[i], max_width=200),
        icon=folium.Icon('orange', icon='star')
    ).add_to(map)

# 다양한 색상 리스트
colors = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0']

# 각 PolyLine에 다른 색상 및 툴팁 적용
color_cycle = cycle(colors)
for i in range(len(location_data) - 1):
    coords = [location_data[i], location_data[i + 1]]
    color = next(color_cycle)
    folium.PolyLine(locations=coords, color=color, weight=5, tooltip=f'Polyline {i+1}').add_to(map)

# 각 좌표에 마커 및 색상 적용
for coords, color in zip(location_data, colors):
    folium.CircleMarker(location=coords, radius=5, color=color, fill=True, fill_color=color).add_to(map)

map
```
